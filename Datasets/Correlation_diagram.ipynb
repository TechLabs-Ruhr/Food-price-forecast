```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

#Source: https://artemoppermann.com/de/explorative-datenanalyse-in-machine-learning/
```

```python
df = pd.read_csv('data_merging.csv', index_col=[0])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
facet = sns.FacetGrid(df, aspect=3)
facet.map(sns.kdeplot,'Precipitation (l/m²)', fill= True)
facet.set(xlim=(0, df['Precipitation (l/m²)'].max()))
facet.add_legend()
plt.show()
```

```python
plt.subplots(figsize=(6,5))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="RdYlGn", cbar=True, linecolor='black', linewidths=0.5)
plt.show()
```
