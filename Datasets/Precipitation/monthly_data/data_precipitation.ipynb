## Formatting the data (air temperature for each month)

```python
import pandas as pd
```

### Import data in .csv format and delete data before 1996

```python
df = pd.read_csv('July.txt', sep=";", header=1) 
#source: https://opendata.dwd.de/climate_environment/CDC/regional_averages_DE/monthly/precipitation/
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.drop(df[df['Jahr']<1996].index)
```

### Merging the columns "Jahr" and "Monat"

```python
df['Time'] = df['Jahr'].astype(str) + '-' + '0' + df["Monat"].astype(str)
df.rename(columns={'Deutschland': 'Precipitation (l/m²)'}, inplace=True)
```

```python
df.head()
```

### Delete unnecassary columns and change the order

```python
df.dtypes
```

```python
columns_to_delete = list(range(0,18))
new_columns_order = ['Time', 'Precipitation (l/m²)']
```

```python
df = df.drop(df.columns[columns_to_delete], axis=1)
df = df.drop('Unnamed: 19', axis=1)
df = df[new_columns_order]
```

```python
df.head()
```

### Saving the data

```python
df.to_csv('July_precipitation.csv')
```
