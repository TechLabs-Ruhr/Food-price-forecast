```python
import pandas as pd
```

```python
df = pd.read_csv('Januar.txt', sep=";", header=1) 
#Quelle: https://opendata.dwd.de/climate_environment/CDC/regional_averages_DE/monthly/air_temperature_mean/
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.drop(df[df['Jahr']<1996].index)
```

```python
df['TIME'] = df['Jahr'].astype(str) + '-' + '0' + df["Monat"].astype(str)
df.rename(columns={'Deutschland': 'Air_Temperature'}, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
columns_to_delete = list(range(0,18))
new_columns_order = ['TIME', 'Air_Temperature']
```

```python
df = df.drop(df.columns[columns_to_delete], axis=1)
df = df.drop('Unnamed: 19', axis=1)
df = df[new_columns_order]
```

```python
df.head()
```

```python
df.to_csv('January_air_temperature.csv')
```
