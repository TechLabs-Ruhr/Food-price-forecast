```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Price Forecasting Datensätze/Milling Wheat N2 Futures Historical Data(1).csv")
```

```python
df = df[['Date', 'Price']]
df = df.rename(columns = {'Price': 'Wheat_Price'})
df
```

Formatting the date column

```python
df['Date'] = pd.to_datetime(df['Date'], format = "%m/%d/%Y")
df['Date'] = df['Date'].dt.strftime('%Y-%m')
df
```

```python
data_merging = pd.read_csv("C:/Price Forecasting Datensätze/data_merging_new.csv")
data_merging = data_merging.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis = 1)
data_merging
```

Merging both DataFrames

```python
data_merging = pd.merge(left = data_merging, right =df, how = 'left')
data_merging
```

Saving the data in a csv file

```python
data_merging.to_csv('data_merging_new_updated.csv')
```
